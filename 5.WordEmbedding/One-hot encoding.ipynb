#1 One hot encoding (딥 러닝을 이용한 자연어 처리 입문)



```
pip install konlpy
```

```
# 코엔엘파이의 Okt 형태소 분석기를 이용해 문장에 대해 토큰화 수행
import konlpy 
from konlpy.tag import Okt  
okt=Okt()  
token=okt.morphs("나는 자연어 처리를 배운다")  
print(token)
```

```
# 각 토큰에 고유한 정수 아이디 부여
word2index={}
for voca in token:
     if voca not in word2index.keys():
       word2index[voca]=len(word2index)
print(word2index)
```

```
# 토큰을 입력하면 해당 토큰에 대한 원-핫 벡터를 만들어내는 함수 정의
def one_hot_encoding(word, word2index):
       one_hot_vector = [0]*(len(word2index))
       index=word2index[word]
       one_hot_vector[index]=1
       return one_hot_vector
```

```
# "자연어"의 인덱스가 2이므로, one-hot 벡터는 인덱스 2의 값이 1이며, 나머지는 0인 벡터를 출력
one_hot_encoding("자연어",word2index)
```

#2 One-hot Encoding example with KERAS

```
#입력문
text="나랑 점심 먹으러 갈래 점심 메뉴는 햄버거 갈래 갈래 햄버거 최고야"
```

```
# 케라스를 이용한 원핫인코딩
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical

text="나랑 점심 먹으러 갈래 점심 메뉴는 햄버거 갈래 갈래 햄버거 최고야"

t = Tokenizer()
t.fit_on_texts([text])
print(t.word_index) # 각 단어에 대한 인코딩 결과 출력.
```

```
# texts_to_sequences()를 통해서 이를 정수시퀀스로 변환가능
# 단어집합 내의 일부 단어들로만 구성된 sub_text를 생성
sub_text="점심 먹으러 갈래 메뉴는 햄버거 최고야"
encoded = t.texts_to_sequences([sub_text])[0]
print(encoded)
```

```
#케라스는 정수 인코딩된 결과로부터 원핫인코딩을 수행하는 to_categorical() 함수 지원
one_hot = to_categorical(encoded)
print(one_hot)
```
