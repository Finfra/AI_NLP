```
x_data=data['string']
y_data=data['label']
```

```

from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from konlpy.tag import *
import numpy as np
import pandas as pd

corpus = x_data
s=" ".join(corpus)

posToUse=["NNP","NNG","MAG","NP","VV","VV+EF","VV+EP",'XSV+EC']

stopwords=["비극","18"]

def getTokens(s):
    global posToUse
    return [ i[0] for i in  Mecab().pos(s) if i[1] in posToUse ]

def getToken(s,pos=posToUse,stopword=stopwords):
    return [ w for w,t in  Mecab().pos(s) if t in pos and w not in stopword ]

vect = CountVectorizer(tokenizer=getToken)
vect.fit(getTokens(s))
print('Dictionary : ',vect.vocabulary_)
train_x_c=vect.transform([" ".join(getTokens(i)) for i in corpus]).toarray()
train_x_c
```

```
# 파워드 바이 rlaghgus51@gmail.com
batch_size = 5
num_classes = 1
epochs = 50

model = Sequential()
model.add(Dense(40, activation='relu', input_shape=(61,)))
#model.add(Dropout(0.2))
model.add(Dense(10, activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(5, activation='relu'))
#model.add(Dropout(0.2))

model.add(Dense(num_classes, activation='sigmoid'))

model.summary()

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```
history = model.fit(train_x_c, y_data,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1
)
```
